```python
import basico
import os
import json
from typing import Dict
import pandas as pd
from IPython.display import display
import vcell_opt
from pathlib import Path
```

```python
optFile = Path(os.getcwd()) / "test_data" / "optproblem.json"
report_file = Path(os.getcwd()) / "test_data" / "optReport.csv"
with open(optFile, "rb") as f_optfile:
    vcell_opt_problem: vcell_opt.OptProblem = vcell_opt.OptProblem.from_json_data(json.load(f_optfile))

if report_file.exists():
    os.remove(report_file)

copasi_model = basico.load_model(vcell_opt_problem.math_model_sbml_contents)
basico.get_parameters()
```

```python
time_course_result = basico.run_time_course(start_time=0, duration=2, use_number=True)

exp_data = vcell_opt.get_reference_data(vcell_opt_problem)

ax = time_course_result.plot(y='Values[s0]')
time_course_result.plot(y='Values[s1]', ax=ax)

exp_data.plot(x='Time', y='Values[s1]', style='o', ax=ax);
```

```python
display(vcell_opt_problem.parameter_description_list)
```

```python
basico.add_experiment('exp1', data=exp_data)
task_settings = basico.get_task_settings('Parameter Estimation')
task_settings['method'] = vcell_opt.get_copasi_opt_method_settings(vcell_opt_problem)
display(task_settings)
basico.set_task_settings('Parameter Estimation', task_settings)

# add report definition and set report file
basico.add_report('parest report', task=basico.T.PARAMETER_ESTIMATION,
                  body=['CN=Root,Vector=TaskList[Parameter Estimation],Problem=Parameter Estimation,Reference=Function Evaluations',
                        '\\\t',
                        'CN=Root,Vector=TaskList[Parameter Estimation],Problem=Parameter Estimation,Reference=Best Value',
                        '\\\t',
                        'CN=Root,Vector=TaskList[Parameter Estimation],Problem=Parameter Estimation,Reference=Best Parameters'
                        ],
                  )
basico.assign_report("parest report", task=basico.T.PARAMETER_ESTIMATION, filename=str(report_file), append=True)

fit_items = vcell_opt.get_fit_parameters(vcell_opt_problem)
display(fit_items)
basico.set_fit_parameters(fit_items)
display(basico.get_fit_parameters())
```

```python
results: pd.DataFrame = basico.run_parameter_estimation(update_model=True)
basico.plot_per_experiment()
```

```python
# read last line in report and see that the parameter values match
iterations = []
objective_functions = []
with open(report_file, "r") as f_reportfile:
    for line in f_reportfile:
        tokens = line.split("\t")
        iterations.append(float(tokens[0]))
        objective_functions.append(float(tokens[1]))

df = pd.DataFrame({"iteration": iterations, "objectFunc": objective_functions})
df.plot(x="iteration", y="objectFunc", logy=True)
```
